```python
def calculate_stats(data, field="B_MM"):
    """Calculate basic statistics for a given field in the data."""
    values = ak.to_numpy(data[field])
    return {
        "mean": np.mean(values),
        "median": np.median(values),
        "std": np.std(values),
        "min": np.min(values),
        "max": np.max(values),
        "count": len(values)
    }

# Print statistics for each dataset
print("B_MM statistics for DD data:")
print(calculate_stats(dd_data))

print("\nB_MM statistics for LL data:")
print(calculate_stats(ll_data))

print("\nB_MM statistics for combined data:")
print(calculate_stats(combined_data))
```

```python
# Cell 9: Fit the B mass distribution with PyROOT (Gaussian + linear background)
import ROOT
from ROOT import TCanvas, TH1F, TF1, gStyle, kRed, kBlue, kGreen, TLegend

def fit_b_mass_pyroot(data, bins=100, range_min=5200, range_max=5400, title=""):
    """
    Fit B mass with a Gaussian signal plus linear background using PyROOT.
    Returns the histogram and fit function.
    """
    # Extract B mass data
    values = ak.to_numpy(data["B_MM"])
    
    # Create and fill histogram
    hist_name = "h_" + title.replace(" ", "_").replace("-", "_")
    hist = TH1F(hist_name, title, bins, range_min, range_max)
    
    for val in values:
        if range_min <= val <= range_max:
            hist.Fill(val)
    
    # Set histogram style
    hist.SetLineWidth(2)
    hist.SetLineColor(kBlue)
    hist.SetMarkerStyle(20)
    hist.SetMarkerSize(0.7)
    hist.GetXaxis().SetTitle("B Mass [MeV/c^{2}]")
    hist.GetYaxis().SetTitle(f"Events / {(range_max-range_min)/bins:.1f} MeV/c^{2}")
    
    # Create fit function (Gaussian + linear background)
    # Parameters: 
    # [0], [1], [2]: Gaussian amplitude, mean, sigma
    # [3], [4]: Linear background slope and intercept
    fit_func = TF1("fit_" + hist_name, 
                  "gaus(0) + pol1(3)", 
                  range_min, range_max)
    
    # Set initial parameter values
    mean_guess = np.mean(values)
    sigma_guess = np.std(values)
    max_bin_content = hist.GetMaximum()
    
    fit_func.SetParameters(
        max_bin_content * 0.8,  # Gaussian amplitude
        mean_guess,             # Gaussian mean
        sigma_guess,            # Gaussian sigma
        0,                      # Linear slope
        max_bin_content * 0.2   # Linear intercept
    )
    
    # Set parameter names for better readability
    fit_func.SetParNames("Amplitude", "Mean", "Sigma", "Slope", "Intercept")
    
    # Define fit range - can be narrower than histogram range
    fit_range_min = range_min
    fit_range_max = range_max
    
    # Perform the fit
    fit_result = hist.Fit(fit_func, "SR", "", fit_range_min, fit_range_max)
    
    # Draw the histogram and fit
    canvas = TCanvas("c_" + hist_name, title, 800, 600)
    gStyle.SetOptStat(0)
    gStyle.SetOptFit(0)  # Don't show fit box, we'll print it ourselves
    
    hist.Draw("E")
    
    # Extract fit components for visualization
    signal_func = TF1("signal_" + hist_name, "gaus", range_min, range_max)
    signal_func.SetParameters(
        fit_func.GetParameter(0),
        fit_func.GetParameter(1),
        fit_func.GetParameter(2)
    )
    signal_func.SetLineColor(kRed)
    signal_func.SetLineStyle(2)
    signal_func.Draw("same")
    
    bkg_func = TF1("bkg_" + hist_name, "pol1", range_min, range_max)
    bkg_func.SetParameters(
        fit_func.GetParameter(3),
        fit_func.GetParameter(4)
    )
    bkg_func.SetLineColor(kGreen+2)
    bkg_func.SetLineStyle(2)
    bkg_func.Draw("same")
    
    # Add legend
    legend = TLegend(0.65, 0.65, 0.89, 0.89)
    legend.AddEntry(hist, "MC Data", "lep")
    legend.AddEntry(fit_func, "Gaussian + Linear", "l")
    legend.AddEntry(signal_func, "Signal Component", "l")
    legend.AddEntry(bkg_func, "Background Component", "l")
    legend.Draw()
    
    canvas.Update()
    canvas.SaveAs(f"{hist_name}.pdf")
    
    # Print fit results
    print(f"Fit results for {title}:")
    print(f"  Signal Parameters:")
    print(f"    Amplitude: {fit_func.GetParameter(0):.1f}  {fit_func.GetParError(0):.1f}")
    print(f"    Mean: {fit_func.GetParameter(1):.2f}  {fit_func.GetParError(1):.2f} MeV/c")
    print(f"    Sigma: {fit_func.GetParameter(2):.2f}  {fit_func.GetParError(2):.2f} MeV/c")
    print(f"  Background Parameters:")
    print(f"    Slope: {fit_func.GetParameter(3):.6f}  {fit_func.GetParError(3):.6f}")
    print(f"    Intercept: {fit_func.GetParameter(4):.1f}  {fit_func.GetParError(4):.1f}")
    print(f"  Fit quality:")
    print(f"    Chi2/NDF: {fit_func.GetChisquare():.1f}/{fit_func.GetNDF()} = {fit_func.GetChisquare()/fit_func.GetNDF():.2f}")
    
    # Calculate signal yield
    bin_width = (range_max - range_min) / bins
    signal_integral = signal_func.Integral(range_min, range_max) / bin_width
    signal_error = signal_integral * (fit_func.GetParError(0) / fit_func.GetParameter(0))
    
    print(f"  Signal yield: {signal_integral:.1f}  {signal_error:.1f}")
    
    return hist, fit_func, canvas

# Apply the PyROOT fit to the combined data
hist, fit_func, canvas = fit_b_mass_pyroot(
    combined_data, 
    bins=100, 
    range_min=5200, 
    range_max=5400, 
    title=f"B Mass Fit - {decay_mode}"
)

# Keep the canvas active (prevents it from being garbage collected)
ROOT.SetOwnership(canvas, False)
```

```python
# Cell 10: Perform a more sophisticated fit using RooFit
from ROOT import RooRealVar, RooDataHist, RooArgList, RooArgSet, RooFit
from ROOT import RooGaussian, RooPolynomial, RooAddPdf, RooDataSet, RooHistPdf
from ROOT import kDashed, kRed, kBlue, kGreen

def fit_with_roofit(data, bins=100, range_min=5200, range_max=5400, name_prefix="roofit"):
    """
    Perform a more sophisticated fit using RooFit with a Gaussian signal plus polynomial background.
    
    Args:
        data: Input data (awkward array)
        bins: Number of bins
        range_min: Lower mass range
        range_max: Upper mass range
        name_prefix: Prefix for output names
    
    Returns:
        RooFitResult and canvas with fit
    """
    # Extract B mass data
    b_mass_values = ak.to_numpy(data["B_MM"])
    
    # Create mass variable
    mass = RooRealVar("mass", "B Mass [MeV/c^{2}]", range_min, range_max)
    
    # Create RooDataSet from numpy array
    data_list = []
    for val in b_mass_values:
        if range_min <= val <= range_max:
            mass.setVal(val)
            data_list.append(mass.Clone())
    
    dataset = RooDataSet("dataset", "B Mass Dataset", RooArgSet(mass))
    for point in data_list:
        dataset.add(RooArgSet(point))
    
    # Create binned version for display
    mass_bins = RooFit.Binning(bins, range_min, range_max)
    data_hist = dataset.binnedClone()
    
    # Define the fit model (Gaussian signal + polynomial background)
    # Signal parameters
    mean = RooRealVar("mean", "Mean", 5280, 5250, 5300)
    sigma = RooRealVar("sigma", "Sigma", 15, 5, 30)
    gauss = RooGaussian("gauss", "Gaussian Signal", mass, mean, sigma)
    
    # Background parameters
    c0 = RooRealVar("c0", "c0", 1.0, -10, 10)
    c1 = RooRealVar("c1", "c1", -0.1, -10, 10)
    bkg = RooPolynomial("bkg", "Background", mass, RooArgList(c0, c1))
    
    # Combined model
    sig_yield = RooRealVar("sig_yield", "Signal Yield", dataset.numEntries()*0.8, 0, dataset.numEntries()*2)
    bkg_yield = RooRealVar("bkg_yield", "Background Yield", dataset.numEntries()*0.2, 0, dataset.numEntries()*2)
    
    model = RooAddPdf("model", "Signal + Background", 
                      RooArgList(gauss, bkg), 
                      RooArgList(sig_yield, bkg_yield))
    
    # Perform the fit
    fit_result = model.fitTo(dataset, RooFit.Save(), RooFit.PrintLevel(-1))
    
    # Create canvas for plotting
    canvas = TCanvas(f"c_{name_prefix}", "RooFit Result", 900, 700)
    canvas.Divide(1, 2)
    canvas.cd(1)
    
    # Create frame for plotting
    frame = mass.frame(RooFit.Title("B Mass Fit with RooFit"))
    
    # Plot data and fit on the frame
    dataset.plotOn(frame, RooFit.Name("data"))
    model.plotOn(frame, RooFit.Name("model"))
    model.plotOn(frame, RooFit.Components("bkg"), RooFit.LineStyle(kDashed), 
                RooFit.LineColor(kGreen+2), RooFit.Name("bkg"))
    model.plotOn(frame, RooFit.Components("gauss"), RooFit.LineStyle(kDashed), 
                RooFit.LineColor(kRed), RooFit.Name("signal"))
    
    # Add fit statistics to the plot
    model.paramOn(frame, RooFit.Layout(0.65, 0.9, 0.9))
    frame.Draw()
    
    # Create the pull distribution plot
    canvas.cd(2)
    pull_frame = mass.frame(RooFit.Title("Pull Distribution"))
    hpull = frame.pullHist()
    pull_frame.addPlotable(hpull, "P")
    pull_frame.Draw()
    
    canvas.Update()
    canvas.SaveAs(f"{name_prefix}_fit.pdf")
    
    # Print fit results
    print(f"\nRooFit results:")
    print(f"  Signal Mean: {mean.getVal():.2f}  {mean.getError():.2f} MeV/c")
    print(f"  Signal Sigma: {sigma.getVal():.2f}  {sigma.getError():.2f} MeV/c")
    print(f"  Signal Yield: {sig_yield.getVal():.1f}  {sig_yield.getError():.1f}")
    print(f"  Background Yield: {bkg_yield.getVal():.1f}  {bkg_yield.getError():.1f}")
    
    return fit_result, canvas, model

# Fit the combined data using RooFit
roofit_result, roofit_canvas, roofit_model = fit_with_roofit(
    combined_data, 
    bins=100, 
    range_min=5200, 
    range_max=5400, 
    name_prefix=f"roofit_{decay_mode}"
)

# Keep the canvas active
ROOT.SetOwnership(roofit_canvas, False)# Cell 11: Plot the fit results together
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages

def plot_fit_summary(dd_data, ll_data, combined_data, decay_mode):
    """
    Create a summary plot of all B mass distributions and fits.
    
    Args:
        dd_data: DD tuple data
        ll_data: LL tuple data
        combined_data: Combined data
        decay_mode: Decay mode name
    """
    # Create a multi-page PDF
    with PdfPages(f"B_mass_summary_{decay_mode}.pdf") as pdf:
        # Plot 1: All three histograms overlaid
        plt.figure(figsize=(10, 8))
        
        # Plot histograms
        plt.hist(ak.to_numpy(dd_data["B_MM"]), bins=100, range=(5200, 5400), 
                 histtype='step', linewidth=2, alpha=0.7, label='DD')
        plt.hist(ak.to_numpy(ll_data["B_MM"]), bins=100, range=(5200, 5400), 
                 histtype='step', linewidth=2, alpha=0.7, label='LL')
        plt.hist(ak.to_numpy(combined_data["B_MM"]), bins=100, range=(5200, 5400), 
                 histtype='step', linewidth=2, alpha=0.7, label='Combined')
        
        plt.xlabel('B Mass [MeV/c]', fontsize=14)
        plt.ylabel('Events / 2 MeV/c', fontsize=14)
        plt.title(f'B Mass Distributions - {decay_mode}', fontsize=16)
        plt.legend(fontsize=12)
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        
        pdf.savefig()
        plt.close()
        
        # Calculate statistics for the summary table
        dd_stats = calculate_stats(dd_data)
        ll_stats = calculate_stats(ll_data)
        combined_stats = calculate_stats(combined_data)
        
        # Create a table of statistics
        fig, ax = plt.subplots(figsize=(10, 6))
        ax.axis('tight')
        ax.axis('off')
        
        data = [
            ['', 'DD', 'LL', 'Combined'],
            ['Events', dd_stats['count'], ll_stats['count'], combined_stats['count']],
            ['Mean [MeV/c]', f"{dd_stats['mean']:.2f}", f"{ll_stats['mean']:.2f}", f"{combined_stats['mean']:.2f}"],
            ['Std Dev [MeV/c]', f"{dd_stats['std']:.2f}", f"{ll_stats['std']:.2f}", f"{combined_stats['std']:.2f}"],
            ['Min [MeV/c]', f"{dd_stats['min']:.2f}", f"{ll_stats['min']:.2f}", f"{combined_stats['min']:.2f}"],
            ['Max [MeV/c]', f"{dd_stats['max']:.2f}", f"{ll_stats['max']:.2f}", f"{combined_stats['max']:.2f}"]
        ]
        
        table = ax.table(cellText=data, loc='center', cellLoc='center')
        table.auto_set_font_size(False)
        table.set_fontsize(12)
        table.scale(1, 1.5)
        
        plt.title(f'B Mass Statistics - {decay_mode}', fontsize=16)
        plt.tight_layout()
        
        pdf.savefig()
        plt.close()
    
    print(f"Summary plots saved to B_mass_summary_{decay_mode}.pdf")

# Create summary plots
plot_fit_summary(dd_data, ll_data, combined_data, decay_mode)
from ROOT import RooRealVar, RooDataHist, RooArgList, RooArgSet, RooFit
from ROOT import RooGaussian, RooPolynomial, RooAddPdf, RooDataSet, RooHistPdf
from ROOT import kDashed, kRed, kBlue, kGreen

def fit_with_roofit(data, bins=100, range_min=5200, range_max=5400, name_prefix="roofit"):
    """
    Perform a more sophisticated fit using RooFit with a Gaussian signal plus polynomial background.
    
    Args:
        data: Input data (awkward array)
        bins: Number of bins
        range_min: Lower mass range
        range_max: Upper mass range
        name_prefix: Prefix for output names
    
    Returns:
        RooFitResult and canvas with fit
    """
    # Extract B mass data
    b_mass_values = ak.to_numpy(data["B_M"])
    
    # Create mass variable
    mass = RooRealVar("mass", "B Mass [MeV/c^{2}]", range_min, range_max)
    
    # Create RooDataSet from numpy array
    data_list = []
    for val in b_mass_values:
        if range_min <= val <= range_max:
            mass.setVal(val)
            data_list.append(mass.Clone())
    
    dataset = RooDataSet("dataset", "B Mass Dataset", RooArgSet(mass))
    for point in data_list:
        dataset.add(RooArgSet(point))
    
    # Create binned version for display
    mass_bins = RooFit.Binning(bins, range_min, range_max)
    data_hist = dataset.binnedClone()
    
    # Define the fit model (Gaussian signal + polynomial background)
    # Signal parameters
    mean = RooRealVar("mean", "Mean", 5280, 5250, 5300)
    sigma = RooRealVar("sigma", "Sigma", 15, 5, 30)
    gauss = RooGaussian("gauss", "Gaussian Signal", mass, mean, sigma)
    
    # Background parameters
    c0 = RooRealVar("c0", "c0", 1.0, -10, 10)
    c1 = RooRealVar("c1", "c1", -0.1, -10, 10)
    bkg = RooPolynomial("bkg", "Background", mass, RooArgList(c0, c1))
    
    # Combined model
    sig_yield = RooRealVar("sig_yield", "Signal Yield", dataset.numEntries()*0.8, 0, dataset.numEntries()*2)
    bkg_yield = RooRealVar("bkg_yield", "Background Yield", dataset.numEntries()*0.2, 0, dataset.numEntries()*2)
    
    model = RooAddPdf("model", "Signal + Background", 
                      RooArgList(gauss, bkg), 
                      RooArgList(sig_yield, bkg_yield))
    
    # Perform the fit
    fit_result = model.fitTo(dataset, RooFit.Save(), RooFit.PrintLevel(-1))
    
    # Create canvas for plotting
    canvas = TCanvas(f"c_{name_prefix}", "RooFit Result", 900, 700)
    canvas.Divide(1, 2)
    canvas.cd(1)
    
    # Create frame for plotting
    frame = mass.frame(RooFit.Title("B Mass Fit with RooFit"))
    
    # Plot data and fit on the frame
    dataset.plotOn(frame, RooFit.Name("data"))
    model.plotOn(frame, RooFit.Name("model"))
    model.plotOn(frame, RooFit.Components("bkg"), RooFit.LineStyle(kDashed), 
                RooFit.LineColor(kGreen+2), RooFit.Name("bkg"))
    model.plotOn(frame, RooFit.Components("gauss"), RooFit.LineStyle(kDashed), 
                RooFit.LineColor(kRed), RooFit.Name("signal"))
    
    # Add fit statistics to the plot
    model.paramOn(frame, RooFit.Layout(0.65, 0.9, 0.9))
    frame.Draw()
    
    # Create the pull distribution plot
    canvas.cd(2)
    pull_frame = mass.frame(RooFit.Title("Pull Distribution"))
    hpull = frame.pullHist()
    pull_frame.addPlotable(hpull, "P")
    pull_frame.Draw()
    
    canvas.Update()
    canvas.SaveAs(f"{name_prefix}_fit.png")
    canvas.SaveAs(f"{name_prefix}_fit.pdf")
    
    # Print fit results
    print(f"\nRooFit results:")
    print(f"  Signal Mean: {mean.getVal():.2f}  {mean.getError():.2f} MeV/c")
    print(f"  Signal Sigma: {sigma.getVal():.2f}  {sigma.getError():.2f} MeV/c")
    print(f"  Signal Yield: {sig_yield.getVal():.1f}  {sig_yield.getError():.1f}")
    print(f"  Background Yield: {bkg_yield.getVal():.1f}  {bkg_yield.getError():.1f}")
    
    return fit_result, canvas, model

# Fit the combined data using RooFit
roofit_result, roofit_canvas, roofit_model = fit_with_roofit(
    combined_data, 
    bins=100, 
    range_min=5200, 
    range_max=5400, 
    name_prefix=f"roofit_{decay_mode}"
)

# Keep the canvas active
ROOT.SetOwnership(roofit_canvas, False)
```
